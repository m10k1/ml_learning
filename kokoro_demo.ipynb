```python
import torch
import soundfile as sf
from kokoro import KPipeline
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
pipeline = KPipeline(lang_code='j')
```

```python
text = '今日の天気は晴れ、もうすぐ春が来そうな雰囲気です'
```

```python
generator = pipeline(text, voice='af_heart', speed=1, split_pattern=r'\n+')
```

```python
from google.colab import files
```

```python
for i, (gs, ps, audio) in enumerate(generator):
  print(f"Index: {i}\n Graphemes: {gs}\nPhonemes: {ps}")

  output_filename = f'output_audio_{i}.wav'
  sf.write(output_filename, audio, 24000)
  print('fAudio saved as {output_filename}')
  files.download(output_filename)
```

```python
from IPython.display import Audio
```

```python
wn = Audio(output_filename, autoplay=True)
display(wn)
```
